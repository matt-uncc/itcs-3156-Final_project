```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "online_payment_fraud.csv"
fraud_df = pd.read_csv(path)
```

```python
fraud_df.head()
```

```python
rng = np.random.RandomState(0)
indices = rng.choice(np.arange(len(fraud_df)), size=500, replace=False)
sns.pairplot(
    data = fraud_df.iloc[indices],
    x_vars = list(fraud_df.drop('isFraud', axis=1).columns),
    y_vars = 'isFraud',
    hue = 'isFraud',
)
plt.show()
```
